1. N개의 랜선을 만드는 것이 목표


2. 이미 자체적으로 가지고 있는 랜선의 갯수는 K개이다.


3. K개의 랜선의 길이는 제각각이다. 하지만 N개의 랜선의 길이는 모두 동일해야한다.


4. 입력 = K개의 랜선과 , 만들고자하는 N개의 랜선값이 입력된다.  
   그뒤로 K줄에 걸쳐서 K개의 랜선의 길이가 입력된다.


5. K = 1< K < 10000 // 1 < N < 1,000,000 // 항상 K <= N


6. 802cm 랜선에서 4개, 743cm 랜선에서 3개, 457cm 랜선에서 2개, 539cm 랜선에서 2개를 잘라내 모두 11개를 만들 수 있다.

```python
K , N = map(int,input().split())
K_len = [0]*K # K개의 랜선을 담을 공간을 생성
for i in range(K):
    K_len[i] = int(input()) # K개의 랜선을 K_len에 담는다.

K_len.sort() # K_len을 가장 작은수부터 나오도록 정렬
count = 0
st_num = 2

while count == 0:
    
    div_num = K_len[0] // st_num # div_num은 내가 K_len에 있는 랜선들을 자르려고하는 길이
    
    for wires in K_len:
        count += wires // div_num # K_len에 있는 랜선들을 div_num길이만큼 잘라서 count변수에 저장
        
    if count < N: # 만약에 count < N 이면 너무 크게 자른거니까... st_num + 1을해서 다시 while문으로...
        count = 0
        st_num += 1

    elif count > N: # count > N 은 이제 너무 잘게 짜른경우니까 여기에서 div_num의 변수를 하나씩 증가시켜보자

        next_div_num = K_len[0] // (st_num - 1)
        # div_num은 너무 잘게 잘려있고, next_div_num은 너무 크게 짤려있으니까 두개의 수 사이에 내가 찾고자하는
        # 값이 존재하니까... 범위를 설정하기 위해서...
        
        for input_div_num in range(div_num,next_div_num): # for문을 돌자
            count = 0 # count에 잘린 갯수를 저장하니까 계속 초기화를 시켜주고
            for wires in K_len:
                count += wires // input_div_num # K_len에 있는 랜선들을 계속 잘라주고 count에 저장
            if (count + 1) == N: # 만약 count+1 이 되면 (내가 자르고자 하는 갯수보다 커지는 순간)
                # count == N 일때의 최대값을 구하라고 했으니까 커지기 직전에있는 길이를 넣으면 되겠지?
                print(input_div_num - 1) # 1 빼준다....
                break
                
# 이코드로 제출을 했는데... 틀렸단다... 뭐가틀린걸까... 흐음
```
